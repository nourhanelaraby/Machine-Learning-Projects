# Get user input with ratings for each product

```python
userInput = [
            {'name':'Amazon Kindle Lighted Leather Cover,Kindle Keyboard', 'rating':5},
            {'name':'Coconut Water Red Tea 16.5 Oz (pack of 12),Amazon Fire Tv', 'rating':3.5},
            {'name':'Certified Refurbished Amazon Fire TV Stick (Previous Generation - 1st),Kindle Paperwhite', 'rating':2},
            {'name':"Amazon - Amazon Tap Portable Bluetooth and Wi-Fi Speaker - Black,Amazon - Amazon Tap Portable Bluetooth and Wi-Fi Speaker - Black", 'rating':5},
            {'name':'HUAWEI mobile', 'rating':4.5}
            ] 
userproducts = pd.DataFrame(userInput)
userproducts
```

```python
#filtering out products by name
inputId = products_df[products_df['name'].isin(userproducts['name'].tolist())]  
inputId    #gbt 5 products
```

```python
inputproducts = pd.merge(inputId, userproducts)    #el common key name 
#Dropping information we won't use from the input dataframe
inputproducts = inputproducts.drop(['categories'], axis=1)
#Final input dataframe
inputproducts
```

# get user input profile

```python
#Filtering out the products  from the input by id
userproducts = productswithcat_df[productswithcat_df['name'].isin(inputproducts['name'].tolist())]
userproducts
```

```python
#Resetting the index to avoid future issues
userproducts = userproducts.reset_index(drop=True)
#Dropping unnecessary columns
usercategoryTable = userproducts.drop(['id', 'name', 'categories'], axis=1)
usercategoryTable    #this table includes the products the user gave rating
```

# get weighted category matrix

```python
inputproducts['rating']
```

```python
#Dot product to get weighted genre matrix
userProfile = usercategoryTable.transpose().dot(inputproducts['rating'])
#The user profile
userProfile
```

```python
productswithcat_df.sample(5)
```

```python
categorytable = productswithcat_df.set_index(productswithcat_df['id'])   #3shan ashwr 3leeh b sohla
#And drop the unnecessary columns
categorytable = categorytable.drop(['name', 'id', 'categories'], axis=1)
categorytable
```

```python
((categorytable*userProfile).sum(axis=1))/(userProfile.sum())
```

```python
#Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((categorytable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df    #dh total score lw eshrtoh 3leeh
```

```python
recommendationTableWeighted_df = (recommendationTable_df.reset_index()).rename(columns={0: 'weightedAvg'})
recommendationTableWeighted_df
```

```python
products_df=products_df[products_df['id'].isin(recommendationTableWeighted_df['id'].tolist())]
```

```python
user_recommendationTable_df = pd.merge(products_df,recommendationTableWeighted_df,on='id',how='inner')   #personalized 3la user
user_recommendationTable_df
```

```python
user_recommendationTable_df=pd.concat([products_df,recommendationTableWeighted_df],axis=1)
user_recommendationTable_df.sample(5)
```
